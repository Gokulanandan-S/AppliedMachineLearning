```python

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
x =data.iloc[:,:-1].values 
y = data.iloc[:,-1].values
```

```python
print('the shape of independent variables ' , x.shape)
print('the shape of dependent variables ' , y.shape)
```

```python
print('the first five data instances of the independent variables ')
x[:5]
```

```python
print('the first five data instances of the dependent variables ')
y[:5]
```

```python
# train test splitting the data set 
from sklearn.model_selection import train_test_split
x_train , x_test , y_tarin , y_test = train_test_split(x,y,test_size = 0.25,random_state= 2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn3_model = KNeighborsClassifier(n_neighbors=3)
knn5_model = KNeighborsClassifier(n_neighbors=5)
knn7_model = KNeighborsClassifier(n_neighbors=7)
```

```python
knn3_model.fit(x_train, y_tarin)
y_pred = knn3_model.predict(x_test)
from sklearn.metrics import accuracy_score,precision_score,recall_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average = 'macro')
recall = recall_score(y_test, y_pred,average = 'macro')
print('the accuracy :',accuracy)
print('the precision score :',precision)
print('the recall score :',recall)
```

```python
knn5_model.fit(x_train, y_tarin)
y_pred = knn5_model.predict(x_test)
from sklearn.metrics import accuracy_score,precision_score,recall_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average = 'macro')
recall = recall_score(y_test, y_pred,average = 'macro')
print('the accuracy :',accuracy)
print('the precision score :',precision)
print('the recall score :',recall)
```

```python
knn7_model.fit(x_train, y_tarin)
y_pred = knn7_model.predict(x_test)
from sklearn.metrics import accuracy_score,precision_score,recall_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average = 'macro')
recall = recall_score(y_test, y_pred,average = 'macro')
print('the accuracy :',accuracy)
print('the precision score :',precision)
print('the recall score :',recall)
```

```python
from sklearn import metrics
y_pred = knn3_model.predict(x_test)
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred)
auc = metrics.roc_auc_score(y_test, y_pred)

print('area under the curve is ',auc)
plt.plot(fpr,tpr,label="3 neighbours , AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

y_pred = knn5_model.predict(x_test)
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred)
auc = metrics.roc_auc_score(y_test, y_pred)

print('area under the curve is ',auc)
plt.plot(fpr,tpr,label="5 neighbours ,AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

y_pred = knn7_model.predict(x_test)
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred)
auc = metrics.roc_auc_score(y_test, y_pred)

print('area under the curve is ',auc)
plt.plot(fpr,tpr,label="7 neighbours , AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')


plt.show()

```
